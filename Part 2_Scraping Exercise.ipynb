**Q1. Write a function that capitalizes the first and fourth letters of a name.**

    name_changer('macdonald') --> MacDonald
    name_changer('Jen') --> 'Name is too short!'

```python
# Define your function here:
def name_changer(name):
    if len(name) >= 4:
        namelist = list(name)
        namelist[0] = namelist[0].upper()
        namelist[3] = namelist[3].upper()
        print("".join(namelist))
    else: 
        print('Name is too short!')

```

```python
# Test your function here:
name_changer('macdonald')
```

```python
# Test your function here:
name_changer('Jen')
```

**Q2. Write a Python function to multiply all the numbers in a list.**

    multiply([1, 2, 3, -4]) --> -24

```python
# Define your function here:

def multiply(list):
    result = 1
    for i in list:
        result = result*i  
    print(result)


```

**Q3. Write a function that takes a string and counts the number of upper case letters, lower case letters, and non-letter characters.**

    up_low_counter("11 aa BB !!")

Output:

    No. of upper case characters :  2
    No. of lower case characters :  2
    No. of non-letter characters :  7

    up_low_counter("The Academy of Motion Pictures Arts and Sciences told members its leaders are upset and outraged by Will Smith's behaviour during the Oscars.")
    
Output:

    No. of upper case characters :  9
    No. of lower case characters :  108
    No. of non-letter case characters :  24

Hint: Two string methods that might prove useful are `.isupper()` and `.islower()`.


```python
# Define your function here:

def up_low_counter(string):
    No_of_lower = 0
    No_of_upper = 0
    No_of_nonletter =0
    for i in string:
        if i.isupper() == True:
            No_of_upper += 1
        elif i.islower() == True:
            No_of_lower += 1
        else: No_of_nonletter +=1
    print("No. of upper case characters : "+ str(No_of_upper) + 
          "\nNo. of lower case characters :" + str(No_of_lower)+
          "\nNo. of non-letter case characters : "+ str(No_of_nonletter))

```

```python
# Test your function here:
up_low_counter("11 aa BB !!")
up_low_counter("The Academy of Motion Pictures Arts and Sciences told members its leaders are upset and outraged by Will Smith's behaviour during the Oscars.")
```

**Q7. Use your program in the previous question to create a function that takes the URL as input and prints  the restaurant's name and address. Name the function YelpScraper. Make sure to handle HTTPError, URLError, and AttributeError exceptions.**

Test the function with the ilili restaurant:

    YelpScraper('https://www.yelp.com/biz/ilili-new-york-2')  
Output:

    Name:   ilili
    Adress: 236 5th Ave New York, NY 10001
and 

    YelpScraper('https://www.yelp.com/biz/nonexisting_page')

Output:

    The restaurant you are looking for does not exist on Yelp!

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from urllib.error import URLError
from bs4 import BeautifulSoup

# Define your function here:
def YelpScraper(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        print('The restaurant you are looking for does not exist on Yelp!')
        return None
    except URLError as e:
        print('The server could not be found!')
        return None
    bs = BeautifulSoup(html.read(), 'html.parser')
    try:
        name = bs.find('h1', {'class': 'css-12dgwvn'}).get_text()
        address1 = bs.find('p', {'class': 'css-r9996t'}).get_text()
        address2 = bs.find('p',{'class':'css-1sb02f4'}).get_text()
    except AttributeError as e:
        print('Tag was not found')
        return None
    else:
        if name == None:
            print ("Page has no restaurant's name")
            return None
        else:
            print("Name:     "+ name)
            print("Address:  "+ address1 +" "+ address2)
```

    Best Supporting Actor :  Troy Kotsur – CODA as Frank Rossi
    Best Supporting Actress :  Ariana DeBose – West Side Story as Anita
    ...

```python

from urllib.request import urlopen
from bs4 import BeautifulSoup
web_address='https://en.wikipedia.org/wiki/94th_Academy_Awards'
html = urlopen(web_address)
bs = BeautifulSoup(html.read(), 'html.parser')
tablelist = bs.find_all('table', {'class': 'wikitable'})
award_list = []
for item in tablelist:
    awards = item.find_all("b")
    award_list.append(awards)
award_list = award_list[0]
for i,k in zip(award_list[0::2],award_list[1::2]):
    print(str(i.get_text()),':',str(k.get_text()))
 
```

```python
# Professor's solution:
url='https://en.wikipedia.org/wiki/94th_Academy_Awards'

def oscar_wiki_scraper(url):
    from urllib.request import urlopen
    from bs4 import BeautifulSoup

    html = urlopen(url)
    bs = BeautifulSoup(html.read(), 'html.parser')

    categories_data = bs.find_all('td',{'style':'vertical-align:top; width:50%;'})
    for category_data in categories_data:
        category_name = category_data.find('div').get_text()
        winner = category_data.li.b.get_text()
        print(category_name, ':',winner) #gives the first <div> tag
```

```python
url_list = ['https://en.wikipedia.org/wiki/94th_Academy_Awards',
            'https://en.wikipedia.org/wiki/93rd_Academy_Awards',
            'https://en.wikipedia.org/wiki/92nd_Academy_Awards']
for url in url_list:
    oscar_wiki_scraper(url)
    print('\n\n------------------------------------------------------------------------------\n\n')
```
