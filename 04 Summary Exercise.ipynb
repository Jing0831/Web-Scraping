**<font color=blue> Exercise 1: Write a program that prints the highlighted section of the description of each gift on [this webpage](http://www.pythonscraping.com/pages/page3.html). The output must be:**

    Now with super-colorful bell peppers!`
    8 entire dolls per set! Octuple the presents!
    Also hand-painted by trained monkeys!
    Or maybe he's only resting?
    Keep your friends guessing!`

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

web_address='https://www.pythonscraping.com/pages/page3.html'
html = urlopen(web_address)
bs = BeautifulSoup(html.read(), 'html.parser')

nameList = bs.find_all('span', {'class': 'excitingNote'})
for name in nameList:
    print(name.get_text())
```

```python
# Here is another solution using keyword input argument of the find_all function. 

from urllib.request import urlopen
from bs4 import BeautifulSoup

web_address='http://www.pythonscraping.com/pages/page3.html'
html = urlopen(web_address)
bs = BeautifulSoup(html.read(), 'html.parser')

descriptionList = bs.find_all(class_ = 'excitingNote')
for description in descriptionList:
    print(description.get_text())
```

**<font color=blue> Exercise 5: Write a program that scrapes the title of the 50 greatest books of all time from [this page](https://thegreatestbooks.org/). The output should be:** </font> 

    In Search of Lost Time 
    Don Quixote
    Ulysses
    The Great Gatsby
    ...

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

web_address='https://thegreatestbooks.org/'
html = urlopen(web_address)
bs = BeautifulSoup(html.read(), 'html.parser')
ListPage1 = bs.find_all('li',{'class':'item pb-3 pt-3 border-bottom'})
booklist = []
for books in ListPage1:
    book_title = books.h4.get_text().split("by")[0].split(".")[1]
    print(book_title)
    booklist.append(book_title)

print(booklist)
```

```python
# Another way:
from urllib.request import urlopen
from bs4 import BeautifulSoup

web_address='https://thegreatestbooks.org/'
html  = urlopen(web_address)
bs    = BeautifulSoup(html.read(), 'html.parser')
body  = bs.find('div',{'class':'list-body'})
books = body.find_all('div',{ 'class':'container'})
for book in books:
    print(book.find_all('a')[0].get_text())
    print('-----')
```

**<font color=blue> Exercise 7: Use [this Craigslist page](https://newyork.craigslist.org/search/lgi/apa?) and practice going up, down, and sidways. Present an example of each.** </font>

```python
#Going down: the <li> tags are children of the <ul> tag

from urllib.request import urlopen
from bs4 import BeautifulSoup

web_address='https://newyork.craigslist.org/search/lgi/apa?'
html = urlopen(web_address)
bs = BeautifulSoup(html.read(), 'html.parser')
giftChildren=bs.find('ul',{'class':'rows'}).children
# note that it's find() not find_all().
for giftChild in giftChildren:
    print(giftChild)
  
```

```python
#Going sideways: Find the silbings of "housing type" in the checkbox list (laundry, parking, and rent period). 

html = urlopen('https://newyork.craigslist.org/search/lgi/apa?')
bs = BeautifulSoup(html, 'html.parser')

for sibling in bs.find('div', {'class':'searchgroup search-group-multi_checkbox'}).div.next_siblings:
    print(sibling)
```
